```python
na_counts = train.isna().sum()
features_with_na = na_counts[na_counts > 0].index.tolist()

print("Features with missing values:")
if features_with_na:
    for feature in features_with_na:
        print(feature)
else:
        print("No missing value")
```

```python
from sklearn.feature_selection import mutual_info_regression as mir 
features = train.drop(columns='tested_positive.4')
top_features = pd.DataFrame(mir(features, train['tested_positive.4']), index = features.columns).sort_values(by=0, ascending = False)
top_features = top_features.rename(columns={0: 'mutual_info'})
display(top_features)
```

```python
selected_features = top_features[:14]
display(train[selected_features.index])
```
